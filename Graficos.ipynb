```python
import numpy as np

dados = np.random.randn(1000)

fig, axs = plt.subplots(1, 3, figsize=(15,5))
axs[0].hist(dados, bins=10, alpha =0.5, color='blue', edgecolor='black')
axs[0].set_title('10 Bins')

axs[1].hist(dados, bins=30, alpha =0.5, color='blue', edgecolor='black')
axs[1].set_title('30 Bins')

axs[2].hist(dados, bins=50, alpha =0.5, color='blue', edgecolor='black')
axs[2].set_title('50 Bins')

for ax in axs:
  ax.set_xlabel('Valores')
  ax.set_ylabel('Frequência')

plt.tight_layout()
plt.show()
```

```python
plt.hist(df['nota'], edgecolor='black',linewidth=1.9,bins=5)
plt.show()
```

```python
#definir a largura da barra
bin_width = 5

#definir onde começa a primeira barra e onde termina a última barra
start_bin = 0
end_bin = 10
min_val = min(df['nota'])
max_val = max(df['nota'])

bins = []
atual = start_bin
while atual <= end_bin:
  bins.append(atual)
  atual += bin_width
plt.hist(df['nota'], edgecolor='black', linewidth=1.2, bins = bins)
plt.show()
```

```python
valores = df['cargo'].value_counts()
labels = valores.index
# unique x index -> unique = verifica diversidade de valores em uma coluna. Index = estrutura que representa linhas e colunas
plt.pie(valores, labels = labels)
plt.title('Gráfico de pizza')
plt.show()
```

```python
valores = df['cargo'].value_counts()
labels = valores.index

plt.pie(valores, labels=labels, autopct='%1.1f%%')
plt.title('Gráfico com porcentagem')
plt.show()
```

```python
explode = [0, 0, 0.1, 0] #'explodir' ou destacar
valores = df['cargo'].value_counts()
labels = valores.index

plt.pie(valores, labels=labels, autopct='%1.1f%%', explode=explode)
plt.title('Gráfico com porcentagem')
plt.show()
```

```python
#Plotar as diferenças entre salários por estados
salario_por_estado = df.groupby('estado')['salario'].mean().reset_index()

plt.figure(figsize=(10,6)) #ajustar o tamanho da figura
plt.bar(salario_por_estado['estado'],salario_por_estado['salario'])
plt.title('Diferenças de salários por estado')
plt.xlabel('Estado')
plt.ylabel('Salário Médio')
plt.xticks(rotation=90) # muda rótulos do eixo x
plt.tight_layout() #ajustar o layout para evitar que os rótulos fiquem cortados
plt.show()
```

```python
#salário de produção de SP e MG
df_producao = df[df['departamento']=='Produção']

#agrupar
salario_por_estado = df_producao.groupby('estado')['salario'].mean().reset_index()

#filtrar apenas SP e MG
df_sp_mg = salario_por_estado[salario_por_estado['estado'].isin(['SP','MG'])]

plt.figure(figsize=(8,6)) #ajustar o tamanho da figura
plt.bar(df_sp_mg['estado'],df_sp_mg['salario'])
plt.title('Salários médios do departamento de Produção em SP e MG')
plt.xlabel('Estado')
plt.ylabel('Salário Médio')
plt.xticks(rotation=90) # muda rótulos do eixo x
plt.tight_layout() #ajustar o layout para evitar que os rótulos fiquem cortados
plt.show()
```

```python
valores = df[['escolaridade', 'nome']].value_counts()
labels = valores.index

# Defina o explode para destacar apenas algumas fatias
explode = [0.1, 0.1, 0.1, 0, 0, 0, 0, 0, 0, 0]

# Plotagem
plt.figure(figsize=(10,8))
plt.pie(valores, labels = labels, explode = explode, autopct='%1.1f%%')
plt.title('Escolaridade dos funcionários')
plt.show()
```
