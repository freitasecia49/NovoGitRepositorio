```python
tamanho = [50, 60, 70, 80]
preco = [100000, 120000, 140000, 160000]

'''np.array(tamanho) = transforma a lista em array numPy, que é o formato que os modelos aceitam.
reshape(-1,1) = transforma o array em um formato de matria com uma coluna, porque o modelo epera esse formato: uma matriz de entrada.
O parâmetro (-1,1) serve para dizer ao numpy: descubra sozinho quantas linhas eu preciso, desde que tenha uma coluna'''
tamanho = np.array(tamanho).reshape(-1, 1)
preco = np.array(preco)

'''LinearRegression() = criar um modelo de regressão linear (uma reta)
.fit(tamanho, preco) = o modelo aprende a relação entre o tamanho de casa e o preço'''

modelo = LinearRegression()
modelo.fit(tamanho, preco)
```
