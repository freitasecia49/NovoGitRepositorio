```python
#bibliotecas
import pandas as pd # manipulação de dados
import numpy as np # operações numéricas e científicas
import matplotlib.pyplot as plt # criação de gráficos
import seaborn as sns # visualização de dados baseados no matplotlib

from sklearn.linear_model import LinearRegression # importar o modelo de Regressão LInear
from sklearn.metrics import mean_squared_error, r2_score # importa métricas de avaliação de modelos
from sklearn.model_selection import train_test_split # importa a função que dividi o conjunto de dados em treino e teste
from sklearn.preprocessing import StandardScaler # importa a classe para escalar os dados, garantindo que todas as variáveis estejam na mesma classe

import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

import scipy.stats as stats


```

```python
tamanho = [50, 60, 70, 80]
preco = [100000 , 120000 , 140000, 160000]

tamanho = np.array(tamanho).reshape(-1,1)
'''
np.array(tamanho) = transforma a lista em array numPy, que é o formato que os modelos aceitam
.reshape(-1, 1) = transforma o array em um formato de matriz com uma coluna,
porque o modelo espera esse formato: uma matriz de entrada.
O parâmetro (-1,1) serve para dizer ao numpy: descubra sozinho quantas linhas eu preciso, desde que
tenha uma coluna
'''
modelo = LinearRegression()
modelo.fit(tamanho, preco)
'''
LinearRegression() = cria um modelo de regressão linear (uma reta)
.fit(tamanho, preco) = o modelo aprende a relação entre o tamanho da casa e o preço
'''
tamanho_para_plotar = np.linspace(tamanho.min(), tamanho.max(), 100).reshape(-1, 1)
preco_previsto = modelo.predict(tamanho_para_plotar)

'''
np.linspace(...): cria 100 valores entre o menor e o maior tamanho da casa. Serve para desenhar a linha do gráfico
(regressão) com maior suavidade

modelo.predict(...): o modelo prevê os preços para esses 100 tamanhos - gera pontos da linha vermelha no gráfico
'''

#Plotar
plt.figure(figsize=(10,4))
plt.scatter(tamanho, preco, color = 'red', label = 'Casas(dados reais)') #gráfico de depressão
plt.plot(tamanho_para_plotar, preco_previsto, color='blue', linewidth=2, label='Linha de regressão')
plt.xlabel('Tamanho da casa (m²)')
plt.ylabel('Preço da casa ($R)')
plt.title('Regressão Linear: Preço vs. Tamanho da casa')
plt.legend()
plt.grid(True)
plt.show()
```

TREINAMENTO MODELO DE REGRESSÃO
1° Preparar os dados
2° Converter os dados em array
3° Formatado para ter uma coluna
4° Instanciei o modelo - iniciada LinearRegression()
5° Treinei o modelo - modelo.fit(tamanho, preco)
6° Predição do modelo - prever os preços para novos tamanhos de casas
7° Visualização - Gráficos


```python
r2 = modelo.score(tamanho, preco) #coeficiente de determinação (R²) -> mede a fração da variância nos dados.
print(f'R²: {r2}')
# significa que o modelo explica perfeitamento os dados,
# r² = próximo de zero, não significa nada
```

```python
#Regressão Linear Simples vs Regressão Linear Múltipla
#RLM: modela a relação entre uma variável dependente e duas ou mais variáveis independentes

'''
Variável independente: é aquela que é manipulada ou controlada para avaliar o seu efeito sobre outra
variável. Ela é independente porque não é afetada por outras variáveis.

Variável dependente: é aquela que é medida ou observada para avaliar como ela muda em resposta
às alterações na variável independente
'''
```

```python
'''
Considerando uma base com dados sobre investimento em propaganda e vendas
repita o processo feito anteriormente
propaganda = [1, 2, 3, 4, 5]
vendas = [1, 2, 4, 5, 8]
'''
```

```python
propaganda = [1, 2, 3, 4, 5]
vendas = [1, 2, 4, 5, 8]

propaganda = np.array(propaganda).reshape(-1,1)

modelo = LinearRegression()
modelo.fit(propaganda, vendas)

propaganda_para_plotar = np.linspace(propaganda.min(), propaganda.max(), 100).reshape(-1, 1)
vendas_previsto = modelo.predict(propaganda_para_plotar)

#Plotar
plt.figure(figsize=(10,4))
plt.scatter(propaganda, vendas, color = 'blue', label = 'Case Vendas') #gráfico de depressão
plt.plot(propaganda_para_plotar, vendas_previsto, color='gray', linewidth=2, label='Linha de regressão')
plt.xlabel('Propaganda ')
plt.ylabel('Vendas ')
plt.title('Regressão Linear: Vendas vs. Propaganda')
plt.legend()
plt.grid(True)
plt.show()
```

```python
r2 = modelo.score(propaganda, vendas) #coeficiente de determinação (R²) -> mede a fração da variância nos dados.
print(f'R²: {r2}')
```

```python
import statsmodels.api as sm # detalhado
x = np.array(propaganda)
#Adicionar uma constante (intercepto) aos dados
x = sm.add_constant(x)

#definir a variável dependente
y = np.array(vendas)

#Criar o modelo
model = sm.OLS(y, x).fit()

#Imprimir o resumo do modelo
print(model.summary())

```
