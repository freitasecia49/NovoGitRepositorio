```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#df = dataframe -> estrtura de dados bidimencional - linhas e colunas
df = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ0PuvujVNx_vgi1rdpA1SlKHK43yvLvXFRcCU7ieQzr27XoBEXl8-yNBwwkxtwd6NAJfrJUmBlg43v/pub?gid=418474837&single=true&output=csv')
display(df)
```

```python
df.groupby('departamento')['departamento'].count()
#groupby(departamento) = agrupa os dados do Dataframe com base nas colunas departamento
#departamento = selecionar especificamento a coluna departamento dentro dos grupos criados
#count = contar o número de elementos de cada grupo formado.
```

```python
df.groupby('departamento')['departamento'].count().rename('Quantidade de funcionários')
```

```python
df.groupby('departamento').agg(qtde=('departamento','count')).sort_values(by='qtde', ascending=False)
#agg = agreggate = agrupar
```

```python
df.groupby('departamento').agg(qtde=('departamento','count')).sort_values(by='qtde', ascending=True)
```

```python
df.rename(columns={'departamento': 'departamentos'})
```

```python
plt.bar(df['departamento'], df['salario']) # plt.bar(df[eixo_x, eixo_y])
plt.show()
```

```python
plt.bar(df['departamento'], df['salario']) # plt.bar(df[eixo_x, eixo_y])
plt.title('Distribuição dos salários por departamento')
plt.show()
```

```python
#distribuição pro escolaridade
explode = [0.1,0,0]
plt.pie(df['escolaridade'].value_counts(), labels=df['escolaridade'].value_counts().index, autopct='%1.1f%%', explode=explode)
plt.title('Distribuição da escolaridade')
plt.show()
```

```python
#verificar gênero
df['sexo'] = ['Masculino','Ferminino','Feminino','Masculino','Ferminino','Masculino','Masculino','Feminino','Masculino','Feminino']
display(df)
```

```python
#ordenar por ordem alfabética
df = df.sort_values(by='nome')
display(df)
```

```python
plt.pie(df['sexo'].value_counts(), labels=df['sexo'].unique(), autopct='%1.1f%%')
plt.title('Gráfico por gênero')
plt.show()
```

QUAL SALÁRIO MÉDIO POR GÊNERO

```python
'''qual salário médio por gênero?'''
df[['departamento','cargo','salario']].groupby(['departamento','cargo']).mean()
salario_medio_por_genero = df.groupby('sexo')['salario'].mean()
display(salario_medio_por_genero)
```

QUAL IDADE MÁXIMA DE CADA DEPARTAMENTO

```python
df[['departamento','idade']].groupby('departamento').mean()
idade_maxima_por_departamento = df.groupby('departamento')['idade'].max()
display(idade_maxima_por_departamento)
```

QUAL É A NOTA MÉDIA POR CIDADE

```python
nota_media_por_cidade = df.groupby('cidade')['nota'].mean()
nota_media_por_cidade = nota_media_por_cidade.round(1)
display(nota_media_por_cidade)
```

QUAL É  A NOTA MÍNIMA POR CIDADE

```python
nota_minima_por_cidade = df.groupby('cidade')['nota'].min()
display(nota_minima_por_cidade)
```

NOTA MÁXIMA POR CIDADE

```python
nota_maxima_por_cidade = df.groupby('cidade')['nota'].max()
display(nota_maxima_por_cidade)
```

AGRUPE O SALÁRIO MÉDIO, A IDADE MÉDIA E A NOTA MÉDIA POR CARGO

```python
salario_medio_por_cargo = df.groupby('cargo')['salario'].mean()
idade_media_por_cargo = df.groupby('cargo')['idade'].mean()
nota_media_por_cargo = df.groupby('cargo')['nota'].mean()
display(salario_medio_por_cargo)
display(idade_media_por_cargo)
display(nota_media_por_cargo)
```

AGRUPE O SALÁRIO MÉDIO, A IDADE MÉDIA E A NOTA MÉDIA POR CARGO segundo exemplo

```python
cargo_media_salario = df[['cargo','salario','idade','nota']].groupby('cargo').mean()
cargo_media_salario = cargo_media_salario.round(2)
display(cargo_media_salario)
```

QUAL É O VALOR DE SALÁRIO QUE SEPARA OS 75% QUE GANHAM MENOS DOS 25% QUE GANAHM MAIS?

```python
separar_salario_por_quantile = df['salario'].quantile([0.25,0.75])
display(separar_salario_por_quantile)
```
